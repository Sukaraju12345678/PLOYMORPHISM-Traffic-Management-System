```python
class TrafficDevice:
    """Parent class for all traffic devices."""
    def activate(self):
        print("Traffic device is now active")
```

```python
class TrafficLight(TrafficDevice):
    """Traffic light implementation."""
    def activate(self):
        print("Traffic Light: Cycling through Red → Yellow → Green")


class SpeedCamera(TrafficDevice):
    """Speed camera implementation."""
    def activate(self):
        print("Speed Camera: Monitoring vehicle speeds and capturing violations")


class PedestrianSignal(TrafficDevice):
    """Pedestrian signal implementation."""
    def activate(self):
        print("Pedestrian Signal: Showing WALK / DON'T WALK indicators")
```

```python
# Create objects and store them in a list
devices = [
    TrafficLight(),
    SpeedCamera(),
    PedestrianSignal()
]

# Activate all devices WITHOUT checking their types
print("=" * 55)
print("SMART TRAFFIC MANAGEMENT SYSTEM")
print("=" * 55)

for device in devices:
    device.activate()

print("=" * 55)
```

```python
class EmergencySiren(TrafficDevice):
    """Emergency siren — added later without modifying the loop!"""
    def activate(self):
        print("Emergency Siren: Flashing lights and sounding alarm!")
```

```python
# Add the new device to the list
devices.append(EmergencySiren())

# The SAME loop now handles the new device automatically!
print("=" * 55)
print("UPDATED SYSTEM (with EmergencySiren)")
print("=" * 55)

for device in devices:
    device.activate()

print("=" * 55)
```
